```python
class TextToSQLBartDataset(Dataset):
    def __init__(self, df, tokenizer, max_source_len=512, max_target_len=256):
        self.df = df.reset_index(drop=True)

        self.tokenizer = tokenizer

        self.max_source_len = max_source_len

        self.max_target_len = max_target_len

    def __len__(self):
        return len(self.df)

    def _build_prompt(self, question, schema):
        return f"Question: {question}\nSchema: {schema}"

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        question = row["question"]
        schema = row["schema"]
        query = row["query"]

        source_text = self._build_prompt(question, schema)
        target_text = query

        source_for_encoder = self.tokenizer(
            source_text,
            max_length=self.max_source_len,
            truncation=True,
            padding="max_length",
            return_tensors="pt",
        )

        target_for_encoder = self.tokenizer(
            target_text,
            max_length=self.max_target_len,
            truncation=True,
            padding="max_length",
            return_tensors="pt",
        )

        input_ids = source_for_encoder["input_ids"].squeeze(0)
        attention_mask = source_for_encoder["attention_mask"].squeeze(0)
        labels = target_for_encoder["input_ids"].squeeze(0)

        labels[labels == self.tokenizer.pad_token_id] = -100

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "labels": labels
        }
```

```python
MODEL_NAME_BART = "facebook/bart-base"

tokenizer_bart = BartTokenizer.from_pretrained(MODEL_NAME_BART)

model_bart = BartForConditionalGeneration.from_pretrained(MODEL_NAME_BART)
model_bart.to(device)
```

```python
MAX_SOURCE_LEN = 512
MAX_TARGET_LEN = 256

dataset_train = TextToSQLBartDataset(df_train, tokenizer_bart, MAX_SOURCE_LEN, MAX_TARGET_LEN)
dataset_dev = TextToSQLBartDataset(df_dev, tokenizer_bart, MAX_SOURCE_LEN, MAX_TARGET_LEN)
dataset_test = TextToSQLBartDataset(df_test, tokenizer_bart, MAX_SOURCE_LEN, MAX_TARGET_LEN)
```

```python
BATCH_SIZE = 4
TRAIN_EPOCHS = 1
LEARNING_RATE = 5e-5

training_args = Seq2SeqTrainingArguments(
    output_dir="./bart_text2sql",
    num_train_epochs=TRAIN_EPOCHS,
    per_device_train_batch_size=BATCH_SIZE,
    per_device_eval_batch_size=BATCH_SIZE,
    learning_rate=LEARNING_RATE,
    weight_decay=0.01,
    warmup_ratio=0.05,
    fp16=False,
    save_total_limit=1,
    logging_steps=100,
    report_to="none"
)
```

```python
def normalize_sql(sql):
    if sql is None:
        return ""

    sql = sql.strip()

    formatted = sqlparse.format(
        sql,
        keyword_case="lower",
        identifier_case="lower",
        strip_comments=True,
        reindent=False
    )

    formatted = re.sub(r"\s+", " ", formatted)
    formatted = formatted.strip()

    if formatted.endswith(";"):
        formatted = formatted[ : -1].strip()

    return formatted

def compute_em_metrics(pred_sql, label_sql):
    if pred_sql is None:
        return 0, 0

    raw_em = int(pred_sql.strip() == label_sql.strip())
    norm_em = int(normalize_sql(pred_sql) == normalize_sql(label_sql))

    return raw_em, norm_em
```

```python
def pred_sql(model, tokenizer, questions, schemas, max_source_len=512, max_target_len=256):
    source_texts = []
    for question, schema in zip(questions, schemas):
        source_texts.append(f"Question: {question}\nSchema: {schema}")

    sources_for_encoder = tokenizer(
        source_texts,
        max_length=max_source_len,
        padding=True,
        truncation=True,
        return_tensors="pt"
    ).to(model.device)

    input_ids = sources_for_encoder["input_ids"]
    attention_mask = sources_for_encoder["attention_mask"]

    with torch.no_grad():
        generated_ids = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            max_length=max_target_len,
            num_beams=4
        )

    pred_sqls = tokenizer.batch_decode(
        generated_ids,
        skip_special_tokens=True
    )

    return pred_sqls
```

```python
def evaluate_bart(model, tokenizer, df, batch_size=8, max_source_len=512, max_target_len=256):
    raw_ems = []
    norm_ems = []

    for start in range(0, len(df), batch_size):
        end = min(start + batch_size, len(df))

        batch = df.iloc[start : end]

        questions = batch["question"].tolist()
        schemas = batch["schema"].tolist()
        queries = batch["query"].tolist()

        pred_sqls = pred_sql(model, tokenizer, questions, schemas, max_source_len, max_target_len)

        for pred, label in zip(pred_sqls, queries):
            raw_em, norm_em = compute_em_metrics(pred, label)

            raw_ems.append(raw_em)
            norm_ems.append(norm_em)

    raw_em_total = float(np.mean(raw_ems))
    norm_em_total = float(np.mean(norm_ems))

    return raw_em_total, norm_em_total
```

```python
class TextToSQLGPT2Dataset(Dataset):
    def __init__(self, df, tokenizer, max_seq_len=512):
        self.df = df.reset_index(drop=True)

        self.tokenizer = tokenizer

        self.max_seq_len = max_seq_len

    def __len__(self):
        return len(self.df)


    def _build_prompt(self, question, schema):
        return f"Question: {question}\nSchema: {schema}\nSQL:"

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        question = row["question"]
        schema = row["schema"]
        query = row["query"]

        prefix = self._build_prompt(question, schema)

        full_text = prefix + " " + query + self.tokenizer.eos_token

        seq_encode = self.tokenizer(
            full_text,
            max_length=self.max_seq_len,
            truncation=True,
            padding="max_length",
            return_tensors="pt",
        )

        prefix_ids = self.tokenizer(
            prefix,
            add_special_tokens=False
        ).input_ids

        input_ids = seq_encode["input_ids"].squeeze(0)
        attention_mask = seq_encode["attention_mask"].squeeze(0)
        labels = input_ids.clone()

        len_prefix = min(len(prefix_ids), self.max_seq_len)
        labels[ : len_prefix] = -100

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "labels": labels,
        }
```

```python
MODEL_NAME_GPT2 = "gpt2"

tokenizer_gpt2 = GPT2Tokenizer.from_pretrained(MODEL_NAME_GPT2)
tokenizer_gpt2.pad_token = tokenizer_gpt2.eos_token

model_gpt2 = GPT2LMHeadModel.from_pretrained(MODEL_NAME_GPT2)
model_gpt2.resize_token_embeddings(len(tokenizer_gpt2))
model_gpt2.to(device)
```

```python
def pred_sql(model, tokenizer, question, schema, max_new_toks=512):
    prefix = f"Question: {question}\nSchema: {schema}\nSQL:"

    prefix_encode = tokenizer(
        prefix,
        add_special_tokens=False,
        return_tensors="pt"
    ).to(model.device)

    input_ids = prefix_encode["input_ids"]
    attention_mask = prefix_encode["attention_mask"]

    with torch.no_grad():
        generated_ids = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            max_new_tokens=max_new_toks,
            num_beams=4,
            pad_token_id=tokenizer.eos_token_id
        )

    continuation_ids = generated_ids[0][input_ids.shape[1] : ]

    pred_sqls = tokenizer.decode(
        continuation_ids,
        skip_special_tokens=True
    )

    return pred_sqls
```

```python
def evaluate_gpt2(model, tokenizer, df, max_new_toks=512):
    raw_ems = []
    norm_ems = []

    for i, row in df.iterrows():
        question = row["question"]
        schema = row["schema"]
        label = row["query"]

        pred = pred_sql(model, tokenizer, question, schema, max_new_toks)

        raw_em, norm_em = compute_em_metrics(pred, label)

        raw_ems.append(raw_em)
        norm_ems.append(norm_em)

    raw_em_total = float(np.mean(raw_ems))
    norm_em_total = float(np.mean(norm_ems))

    return raw_em_total, norm_em_total
```
